```python
import numpy as np
import pandas as pd
```

```python
file_path="../data/raw/online_retail.xlsx"
df_2009=pd.read_excel(file_path,sheet_name="Year 2009-2010")
df_2010=pd.read_excel(file_path,sheet_name="Year 2010-2011")
df=pd.concat([df_2009,df_2010],ignore_index=True)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns={
    'Invoice': 'InvoiceNo',
    'Price': 'UnitPrice',
    'Customer ID': 'CustomerID'
})
```

```python
#Create Revenue
df['Revenue'] = df['Quantity'] * df['UnitPrice']

```

```python
df = df.dropna(subset=['CustomerID'])
df['CustomerID'] = df['CustomerID'].astype(int)
```

```python
df.iloc[1500]
```

```python
df= df[
    (df['Quantity'] > 0) &
    (df['UnitPrice'] > 0)
]
```

```python
df.dtypes
```

```python
df['InvoiceNo'] = df['InvoiceNo'].astype(str)
```

```python
df.dtypes
```

```python
df.to_csv("../data/processed/retail_cleaned.csv", index=False)
```
